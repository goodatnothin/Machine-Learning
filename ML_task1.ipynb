```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
mean = 5; variance = 1 ; std = np.sqrt(variance);
x = np.linspace(mean - 5*std, mean + 5*std, 11)
plt.plot(x,norm.pdf(x,mean,std))
plt.plot(x,norm.pdf(x,mean,std),'1')
plt.show()
```
